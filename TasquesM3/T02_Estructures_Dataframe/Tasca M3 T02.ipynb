### - Anàlisi estadístic del més rellevant

```python
#Mitjana del temps total de vol, en h:m:s.

def passMinToString(mean):
    hour = int(mean/60)
    rest = mean%60
    sec = int((rest-int(rest))*60) 
    hour = str(hour)
    rest = str(int(rest))
    sec = str(sec)
    return hour + "h " + rest + "m " + sec + "s"

time = passMinToString(dfDelayedFlights["total_flight_time_min"].mean())
print("Mitjana de temps total de vol: " + time)
```

```python
#Mitjana del temps en l'aire, en h:m:s.

def passMinToString(mean):
    hour = int(mean/60)
    rest = mean%60
    sec = int((rest-int(rest))*60) 
    hour = str(hour)
    rest = str(int(rest))
    sec = str(sec)
    return hour + "h " + rest + "m " + sec + "s"
        
time = passMinToString(dfDelayedFlights["air_time_min"].mean())
print("Mitjana de temps en l'aire: " + time)
```

```python
#Distància mitjana de vol, en km.

def mileToKm(meanDistance):
    km = meanDistance*1.609
    return round(km, 2)

distance = mileToKm(dfDelayedFlights["distance_miles"].mean())
print("Distància mitjana de vol: ",distance, "km")
```

```python
#Percentatges de vols cancel·lats i desviats.

#Converteixo unes columnes amb int64 a bool
dfDelayedFlights = dfDelayedFlights.astype({"cancelled": bool, "diverted": bool}) 

#Comptar el nombre de vols cancel·lats
dfDelayedFlights.cancelled.value_counts()


def percentTrues(column):
    trues = column.value_counts()[True]
    total = len(column)
    return round(trues/total*100,2)

percent = percentTrues(dfDelayedFlights["cancelled"])
print("El percentatge de vols cancel·lats és", percent ,"%")   


#Comptar el nombre de vols desviats
dfDelayedFlights.diverted.value_counts()

percent = percentTrues(dfDelayedFlights["diverted"])
print("El percentatge de vols desviats és", percent, "%")
```

```python
#Nombre de vols fets per companyia.

flightsPerCo = dfDelayedFlights.code_company.value_counts()
print(flightsPerCo)
```

```python
#Nombre de vols per dia de la setmana.

#Codi per què surtin els dies de la setmana en el df:
dictDays = {1:"Dilluns", 2:"Dimarts", 3:"Dimecres", 4:"Dijous", 5:"Divendres", 6:"Dissabte", 7:"Diumenge"}
dfDelayedFlights["day_of_week"] = dfDelayedFlights["day_of_week"].map(dictDays)

flightsDaysWeek = dfDelayedFlights.day_of_week.value_counts()
print(flightsDaysWeek)
```

```python
#Percentatge de vols per dia de la setmana.

total = len(dfDelayedFlights["day_of_week"])
percentFlightsDaysWeek = flightsDaysWeek.apply(lambda x : str(round(x*100/total, 1)) + "%")
print(percentFlightsDaysWeek)
```

### - Crea columnes noves (distància en km, velocitat mitjana en km·h^-1, si arriba tard o no)

```python
#Canvio la unitat de distància de milla a km i creo la columna distància en km
dfDelayedFlights["distance_km"] = dfDelayedFlights["distance_miles"].apply(mileToKm)
dfDelayedFlights["average_speed_km_h_1"] = round(dfDelayedFlights["distance_km"] / dfDelayedFlights["air_time_min"]*60, 1)

dfDelayedFlights.head()
```

```python
#Considero que a partir de 15 min, un vol arriba tard
delayLimit = 15
dfDelayedFlights["late"] = dfDelayedFlights["arrive_delay_min"].apply(lambda x: True if x > delayLimit else False)

dfDelayedFlights.head()
```

### - Fer una taula amb les aerolínies amb més retards acumulats

```python
dfFlightsForCo = dfDelayedFlights.groupby(["code_company"])[["late"]].count().sort_values(by=["late"], ascending=False)
print(dfFlightsForCo)
```

```python
#Els 10 vols més endarrerits
dfLateFlights = dfDelayedFlights.nlargest(f, "arrive_delay_min")
dfLateFlights[["flight_num", "arrive_delay_min"]]
```
