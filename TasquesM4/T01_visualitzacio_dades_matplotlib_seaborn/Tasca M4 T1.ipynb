## Exercici 1
Descarrega el dataset adjunt preu lloguer per trimestre i barri.csv extret de la web OpenDataBCN i resumeix-lo estadísticament i gràficament. Crea almenys una visualització per:
- Una variable categòrica (Districte o Barri)
- Una variable numèrica (Preu)
- Una variable numèrica i una categòrica (Districte i Preu)
- Dues variables numèriques (Any o Trimestre i Preu)
- Tres variables (Barri o Districte, Trimestre i Preu)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dfRental = pd.read_csv("/Formación/2022 BarcelonaActiva/CursDSOnline/sprint04VisualitzacioGraficaDades/preu lloguer per trimestre i barri.csv")
dfRental.info()
dfRental.head()
```

### - Resum estadístic del dataset

```python
#Modifico el nom de columnes per treballar més còmode
dfRental.columns = [col.lower() for col in dfRental]
dfRental.columns
```

```python
#Converteixo la columna preu a float64 per fer el resum estadístic
dfRental["preu"] = pd.to_numeric(dfRental["preu"], errors = "coerce")
dfRental.info()
```

```python
#Resum estadístic del dataset
dfRental.describe()
```

### - Resum gràfic del dataset
#### Visualització d'una variable categòrica (barri)

```python
#Visualització del preu del lloguer mensual per barris
sns.set_style("whitegrid")
sns.set_palette('deep')
plt.figure(figsize =(10, 14))
sns.barplot(x='preu', y='nom_barri', data=dfRental, ci=None)
plt.title("Barplot de lloguer mensual per barris", fontsize=14)

#Guardo el gràfic en un fitxer amb format .PNG (exercici 2)
plt.savefig("fig1BplotLloguerPerBarri.png")
```

```python
#Visualització del preu del lloguer mensual per barri
sns.set_style("darkgrid")
plt.figure(figsize =(14, 10))
g = sns.stripplot(x="nom_barri", y="preu", data=dfRental)
plt.title("Stripplot entre lloguer mensual i barris", fontsize=18, color="#6146C1", alpha=1)
plt.xlabel('Barri')
plt.ylabel('Lloguer mensual mitjà / euros')
plt.xticks(rotation=60)

#Guardo el gràfic en un fitxer amb format .PNG (exercici 2)
plt.savefig("fig2StplotPreuPerBarris.png")
```

#### Visualització d'una variable numèrica (preu del lloguer)

```python
#Histograma

#Creo la figura
plt.figure(figsize =(8, 8))

sns.histplot(data=dfRental, x='preu')
plt.title("Histograma entre nombre de barris i el lloguer", fontsize=16, color="#6146C5", alpha=1)

#Guardo el gràfic en un fitxer amb format .PNG (exercici 2)
plt.savefig("fig3HsplotNombreBarrisPerLloguer.png")
```

#### Visualització d'una variable numèrica i una categòrica (Districte i Preu)


```python
#Boxplot
plt.figure(figsize =(13, 8))
sns.boxplot(data=dfRental, y='preu', x='nom_districte')
plt.title("Boxplot entre preu de lloguer i districte", fontsize=16, color="#6146C5", alpha=1)

plt.savefig("fig4BoxplotLloguerPerDistricte.png")
```

```python
#Barplot
plt.figure(figsize =(14, 6))
sns.barplot(x='nom_districte', y='preu', data=dfRental)
plt.title("Barplot entre preu de lloguer i districte", fontsize=18, color="#6146C5", alpha=1)

plt.savefig("fig5BarplotLloguerPerDistricte.png")
```

#### Visualització de dues variables numèriques (Any o Trimestre i Preu)

```python
#Scatterplot entre preu de lloguer i any 2021
sns.set_style("darkgrid")
plt.figure(figsize =(6, 6))
sns.scatterplot(data=dfRental, x='any', y='preu')
plt.title("Scatterplot entre preu i any", fontsize=20, color='#6146C6', alpha=1)
plt.xlabel("Any")
plt.ylabel("Preu del lloguer mensual / euros")
plt.xlim(2020.5, 2021.5)
plt.ylim(0, 2000)

#Guardo el gràfic en un fitxer amb format .PNG (exercici 2)
plt.savefig("fig6ScplotLloguerPerAny.png")
```

```python
#Scatterplot entre preu de lloguer i trimestres del 2021
plt.figure(figsize =(5, 8))
sns.scatterplot(data=dfRental, x='trimestre', y='preu')
plt.title("Scatterplot entre lloguer mensual i trimestres 2021", fontsize=16, color='#6146C6', alpha=1)
plt.xlabel("Trimestre del 2021")
plt.ylabel("Preu de lloguer mensual / euros")

plt.savefig("fig7ScplotLloguerPerTrimestre.png")
```

#### Visualització de tres variables numèriques (Barri o Districte, Trimestre i Preu)

```python
#Scatterplot entre 3 variables
sns.set_theme(style="whitegrid")
sns.relplot(x="trimestre", y="preu", hue="codi_districte",
            alpha=.5, palette="deep",
            height=6, data=dfRental)
plt.title("Scatterplot entre lloguer, trimestres 2021 i districtes", fontsize=16, color='#6146C6', alpha=1)

plt.savefig("fig8ScplotLloguerTrimestreCodiDistricte.png")
```

```python
#Dibuixo el pairplot entre tres variables 
data = dfRental[["preu", "trimestre", "codi_districte"]]
titol = "Taula de relacions amb scatterplot entre les lloguer, T i codi districte"
plt.figure(figsize =(15, 13))
axis = sns.pairplot(data)
plt.title(titol, fontsize=20, color='#6146C6', alpha=0.8, pad=350)
plt.figtext(1, 0, titol, fontsize=12, ha='center')
plt.show()
```
