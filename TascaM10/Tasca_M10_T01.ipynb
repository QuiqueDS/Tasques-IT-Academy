## a.- Web scraping amb BeutifulSoup 

```python
from bs4 import BeautifulSoup
import requests
```

```python
# a.- Obtenció de l'autor de la web http://quotes.toscrape.com/  

url = "http://quotes.toscrape.com/"
result = requests.get(url)
content = result.text

# Obtenció del codi HTML de la pàgina web
soup = BeautifulSoup(content, "lxml")     # Parser "lxml" permet obtenir el codi HTML de content
print(soup.prettify())
```

```python
box = soup.find("div", class_="quote")  # class_ per diferenciar l'argument de les classes usades en python
author = box.find("small", class_="author").get_text()
print(author)
```

```python
# Creació arxiu TXT
#with open("einstein.txt", "w") as file:    # "w"-mode escriptura, també hi ha mode lectura
with open(f"{author}.txt", "w") as file: 
    file.write(author)    
```

```python
# b.- Obtenció d'una imatge de la web https://www.bolsamadrid.es/

url2 = "https://www.bolsamadrid.es/"
response = requests.get(url2)
content = response.content
soup = BeautifulSoup(content, "html.parser")  
print(soup.prettify())
```

```python
banner = soup.find("span", class_="section-banners-element")  # class_ per diferenciar l'argument de les classes usades en python
src = banner.find("img", class_="object-fit").get("src")  # Obtenció src de la imatge (URL)
print(src)    

# Creació arxiu PNG
img = requests.get('https://www.bolsasymercados.es'+src)
with open("imagen_bdm.png", "wb") as file:
    file.write(img.content)
```
